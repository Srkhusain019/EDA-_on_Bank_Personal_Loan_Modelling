##### Q17. Give the IQR of all the variables which are quantitative and continuous.

```python
#Answer
#Countinuous and Qantative Variable only in out Data set: - CCAvg,Age,Experience,Income,Mortgage
A = df['CCAvg']
A.quantile([.25,.5,.75])
IQR = A.quantile(.75)-A.quantile(.25)
print('CCAvg_IQR',IQR)

B = df['Age']
B.quantile([.25,.5,.75])
IQR = B.quantile(.75)-B.quantile(.25)
print('Age_IQR',IQR)

C = df['Experience']
C.quantile([.25,.5,.75])
IQR = C.quantile(.75)-C.quantile(.25)
print('Experience_IQR',IQR)

D = df['Income']
D.quantile([.25,.5,.75])
IQR = D.quantile(.75)-D.quantile(.25)
print('Income_IQR',IQR)

E = df['Mortgage']
E.quantile([.25,.5,.75])
IQR = E.quantile(.75)-E.quantile(.25)
print('Mortgage_IQR',IQR)
```

##### 20. Using the z-score of the income variable, find out the number of observations outside the +-3.

```python
Income_zscore = (df['Income'] - df['Income'].mean()) / df['Income'].std()


# Find the number of observations outside the 3 range
outliers = df[(df['Income_zscore'] < -3) | (df['Income_zscore'] > 3)]
num_outliers = outliers.shape[0]
print(num_outliers)
```
